# Pipeline run

```python
from neandertools import AsteroidCutoutPipeline
```

```python
pipeline = AsteroidCutoutPipeline(
      target="2024 TN57",
      target_type="designation",   # for provisional designations
      start="2024-12-01",
      end="2024-12-30",
      dr="dp1",
      collection="LSSTComCam/DP1",
      bands=["u", "g", "r", "i", "z", "y"],   # single band for cleaner GIF
      step="12h",                   # finer ephemeris grid
      cutout_size=150,             # larger cutouts
      polygon_widening_arcsec=2, # wider search area
  )

gif_path = pipeline.run(
    output_path="asteroid.gif",
    frame_duration_ms=500,       # slower animation
    warp_common_grid=True,       # stable background
    show_ne_indicator=True,      # N/E arrows
  )
```
